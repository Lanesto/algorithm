```python
"""
Version 1: TIMEOUT
- used methods: DP

"""

cache = [None] * (1 << 14) * 20 * 2


# Returns number of possible candidates as
# past egg price
def solution(egg_price, multiples):
    # Convert variables to more handy type
    egg_price = [int(c) for c in str(egg_price)]
    n = len(egg_price)
    digits = sorted(egg_price)

    # Initialize cache
    size = (1 << n) * multiples * 2
    cache[:size] = [None] * size

    def solve(position, taken, mod, less):
        # Used all digits
        if position == n:
            # :less is necessary to exclude :egg_price itself
            return 1 if less and (mod == 0) else 0

        # Access cache
        c_idx = taken + (1 << n) * (mod + multiples * int(less))
        ret = cache[c_idx]
        if ret is not None:
            return ret

        ret = 0
        for next in range(n):
            if not bool(taken & (1 << next)):
                # This condition indicates all numbers made this after
                # picking :next is bigger than current egg price
                if not less and digits[next] > egg_price[position]:
                    continue

                # Skipping duplicant permutations
                if next > 0 \
                   and digits[next-1] == digits[next] \
                   and not bool(taken & (1 << (next - 1))):
                    continue

                ret += solve(
                    position + 1,
                    taken | (1 << next),
                    (mod * 10 + digits[next]) % multiples,
                    less or (digits[next] < egg_price[position])
                ) % 1000000007

        # Store to cache and return
        cache[c_idx] = ret
        return ret

    result = solve(0, 0, 0, False)
    return result


# Main I/O part
def main(rl):
    c = int(rl())
    for _ in range(c):
        e, m = map(int, rl().split())
        result = solution(e, m)
        print(result)


# Additional codes to simulate I/O
try:
    import IPython
except ImportError as _:
    # Submit env
    import sys
    main(sys.stdin.readline)
else:
    from helpers import runner
    # IPython env
    runner.run(main)
```
